```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
df=pd.read_csv("penguins.csv")
df
```

```python
# Remove unnamed column

if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

```python

# Target Variable Distribution (e.g., 'species')
plt.figure(figsize=(6, 4))
sns.countplot(x='species', data=df)
plt.title('Penguin Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python

# Distribution of flipper length
plt.figure(figsize=(6, 4))
sns.histplot(df['flipper_length_mm'], kde=True, bins=15)
plt.title('Flipper Length Distribution')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Body Mass Distribution
plt.figure(figsize=(6, 4))
sns.histplot(df['body_mass_g'], kde=True, bins=20)
plt.title('Body Mass Distribution')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')
plt.show()
```

```python
#comparing body mass across species
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='body_mass_g', data=df)
plt.title('Body Mass vs Penguin Species')
plt.show()
```

```python
#Boxplot of Flipper Length by Species
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='flipper_length_mm', data=df)
plt.title('Flipper Length vs Penguin Species')
plt.show()
```

```python
#Count Plot for Sex split by Species
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', data=df, hue='species')
plt.title('Penguin Count by Sex and Species')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='island', data=df, hue='species')
plt.title('Penguin Species Count by Island')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap - Penguins')
plt.show()
```
