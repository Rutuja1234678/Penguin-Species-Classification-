```python
import pandas as pd
# load the dataset
df=pd.read_csv("penguins.csv")
df
```

```python
df.info()
```

```python
#check total missing values
print(df.isnull().sum())
```

```python
# Define numerical columns (can vary slightly depending on usage)
num_cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

# Fill numerical columns with the median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
# Define categorical columns
cat_cols = ['species', 'island', 'sex']

# Fill categorical columns with the mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#check total missing values
print(df.isnull().sum())
```

```python
# check and convert data types (especially for categorical)
df['flipper_length_mm'] = df['flipper_length_mm'].astype(int)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode 'sex' column
le_sex = LabelEncoder()
df['sex_encoded'] = le_sex.fit_transform(df['sex'])

# Drop the original 'sex' column
df.drop('sex', axis=1, inplace=True)

# Show encoding mapping
sex_mapping = dict(zip(le_sex.classes_, le_sex.transform(le_sex.classes_)))
print("Encoding Map:", sex_mapping)
```

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```
